# DATA INCUBATOR plot #2

In this plot, the mean contaminant concentrations for PM 2.5 and O3 are compared for all studied 150 air quality stations in New York City. From the studied contaminants PM 2.5 and O3 are chosen given their distinct characteristics. PM 2.5 peaks in dense urban environments while O3 shows highest concentrations in suburban areas. 

```python
%matplotlib inline
import geopandas
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
import pandas as pd
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

DATA LOADING

```python
NY = geopandas.read_file("geo_export_124bbdfb-9b30-4597-bd42-dc7ac7c1db44.shp")
```

```python
NY.plot();
```

```python
df1 = pd.read_csv("plot2data.csv")
longitude = df1.longitude
latitude = df1.latitude
meanpersitePM = df1.meanpersitePM
meanpersiteO3 = df1.meanpersiteO3

df1.head(5)
```

DATA PLOTTING

```python
fig, ax= plt.subplots(2,figsize=(20,14))
ax[0]=NY.plot(axes=ax[0],color='white', edgecolor='black')
im1=ax[0].scatter(longitude,latitude, c=meanpersitePM,label='Stations')
divider = make_axes_locatable(ax[0])
cax = divider.append_axes('right', size='5%', pad=0.05)
cb0=fig.colorbar(im1, cax=cax, orientation='vertical')
cb0.set_label('PM2.5 µg/m³')
ax[0].set_ylabel('Latitude')
ax[0].set_xlabel('Longitude')

ax[1]=NY.plot(axes=ax[1],color='white', edgecolor='black')
im2=ax[1].scatter(longitude,latitude, c=meanpersiteO3,label='Stations')
divider = make_axes_locatable(ax[1])
caxb = divider.append_axes('right', size='5%', pad=0.05)
cb1=fig.colorbar(im2, cax=caxb, orientation='vertical')
cb1.set_label('O3 ppm')
ax[1].set_ylabel('Latitude')
ax[1].set_xlabel('Longitude')
ax[0].title.set_text('NYC Mean PM2.5 2008-2016')
ax[1].title.set_text('NYC Mean O3 2008-2016')
plt.rcParams.update({'font.size': 14})
```
